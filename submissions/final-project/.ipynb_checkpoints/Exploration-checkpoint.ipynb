```python
def evaluate_clusters(X,max_clusters):
    error = np.zeros(max_clusters+1)
    error[0] = 0;
    for k in range(1,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(X)
        error[k] = kmeans.inertia_

    plt.plot(range(1,len(error)),error[1:])
    plt.xlabel('Number of clusters')
    plt.ylabel('Error')
```

```python
def analyze_lyrics(song, artist):
    lyrics = []
    filename = os.path.join('data/lyrics/' + song.replace('/','') + '-' + artist + '.csv')
    if os.path.isfile(filename):
        f = open(filename)
        raw = f.read().lower().replace(',', '').replace('feat.','').split(';')

        for line in raw:
            lyrics += line.split()

        lyrics_text = nltk.Text(lyrics)
        lyrics_blob = TextBlob(' '.join(lyrics))

        unique = len(set(lyrics_text))
        uniqueness = unique/(len(lyrics_text)*1.0)
        #print "Unique: " + str(unique)
        #print "Uniqueness: " + str(uniqueness)
        fdist1 = nltk.FreqDist(lyrics_text)
        top_words = fdist1.most_common(10)
        #print top_words
        sentiment = lyrics_blob.sentiment.polarity
        happy = False
        if sentiment > 0:
            happy = True
        #print type(sentiment)
        #print sentiment
        return {'unique': unique, 'uniqueness': uniqueness, 'top_words': top_words, 'polarity': sentiment, 'happy': happy}
```

## Hypothesis 1:      <p/>Songs that contain fewer unique words will chart more successfully than songs with greater amounts of unique words. 

Since we corrected all the song names at the beginning to display special characters correctly, we need to get the raw version of the text again in order to process the song names when we reference the lyrics files scraped from Genius. 

```python
song_charts_unmod = pd.read_csv('data/chart-info-addt.csv', parse_dates=['Debut', 'peak_week'], encoding='latin-1')
print song_charts.head(1)
song_charts_unmod.columns = [label.lower() for label in song_charts_unmod.columns]
song_charts_unmod = song_charts_unmod[['song','artist','featuring', 'duration', 'peak']]
song_charts_unmod
```

```python
songs_artists = song_charts_unmod[['song', 'artist']]
lexical_analysis_data = {}
for ind, row in songs_artists.iterrows():
    song = row['song']
    artist = row['artist']
    print song + ' ' + artist
    stats = analyze_lyrics(song, artist)
    if stats is not None:
        lexical_analysis_data[song + '-' + artist] = stats
```

```python
print len(lexical_analysis_data)

unique_count_only = [lexical_analysis_data[key]['unique'] for key in lexical_analysis_data.keys()]
unique_count_only_sorted = np.sort(unique_count_only)[::-1]
unique_count_only_sorted[:100]

too_big = [key for key in lexical_analysis_data.keys() if lexical_analysis_data[key]['unique'] > 600]
reduced_lex_an_data = {}
for key in lexical_analysis_data:
    if key not in too_big:
        reduced_lex_an_data[key] = lexical_analysis_data[key]
        
happy_only = [[key.split('-')[0], key.split('-')[1], reduced_lex_an_data[key]['happy']] for key in reduced_lex_an_data.keys()]
happy_only = pd.DataFrame(happy_only)
happy_only.columns = ['song', 'artist', 'happy']

unique_count_only = [[key.split('-')[0], key.split('-')[1], reduced_lex_an_data[key]['unique']] for key in reduced_lex_an_data.keys()]
unique_count_only = pd.DataFrame(unique_count_only)
unique_count_only.columns = ['song', 'artist', 'unique']
print unique_count_only.head()

song_charts_unmod = pd.merge(song_charts_unmod, happy_only, on=['song', 'artist'])
song_charts_unmod = pd.merge(song_charts_unmod, unique_count_only, on=['song', 'artist'])
```

## Clustering Lyrics

```python
def get_lyrics(song, artist):
    filename = os.path.join('data/lyrics/' + song.replace('/','') + '-' + artist + '.csv')
    if os.path.isfile(filename):
        f = open(filename)
        raw = f.read().lower().replace(',', '').replace('feat.','').replace(';', ' ').replace('[','').replace(']', '').replace('chorus','').replace('pre-hook','').replace('pre','').replace('hook','').replace('verse','').replace('produced by','').replace('\'m',' am').replace('n\'t', ' cannot').replace('\'ll', 'will').replace('\'ve', 'have').replace(':','').replace('\"','').replace('\'s','').replace('ai','')

        return raw
```

```python
songs = []
artists = []
lyrics = []
for ind, row in songs_artists.iterrows():
    song = row['song']
    artist = row['artist']
    lyric = get_lyrics(song, artist)
    if lyric is not None:
        songs += [song]
        artists += [artist]
        lyrics += [lyric]

print lyrics[0]
```

```python
stemmer = SnowballStemmer("english")
def tokenize_and_stem(text):
    # first tokenize by sentence, then by word to ensure that punctuation is caught as it's own token
    tokens = [word for sent in nltk.sent_tokenize(text) for word in nltk.word_tokenize(sent)]
    filtered_tokens = []
    # filter out any tokens not containing letters (e.g., numeric tokens, raw punctuation)
    for token in tokens:
        if re.search('[a-zA-Z]', token):
            filtered_tokens.append(token)
    stems = [stemmer.stem(t) for t in filtered_tokens]
    return stems


def tokenize_only(text):
    # first tokenize by sentence, then by word to ensure that punctuation is caught as it's own token
    tokens = [word.lower() for sent in nltk.sent_tokenize(text) for word in nltk.word_tokenize(sent)]
    filtered_tokens = []
    # filter out any tokens not containing letters (e.g., numeric tokens, raw punctuation)
    for token in tokens:
        if re.search('[a-zA-Z]', token):
            filtered_tokens.append(token)
    return filtered_tokens

totalvocab_stemmed = []
totalvocab_tokenized = []
for i in lyrics:
    allwords_stemmed = tokenize_and_stem(i) #for each item in 'synopses', tokenize/stem
    totalvocab_stemmed.extend(allwords_stemmed) #extend the 'totalvocab_stemmed' list
    
    allwords_tokenized = tokenize_only(i)
    totalvocab_tokenized.extend(allwords_tokenized)
    
vocab_frame = pd.DataFrame({'words': totalvocab_tokenized}, index = totalvocab_stemmed)
print 'there are ' + str(vocab_frame.shape[0]) + ' items in vocab_frame'
```

## Overall Song Features Effect on Chart Performance

I decided instead of looking at the features in isolation, I would run a linear regression model to see if there were any correlations between them and a rating I generated to determine the success of a song based on weighing the duration on the chart and peak position. Extra emphasis was given to songs that remained on the chart passed the 50 week average and peaked in the Top 50, Top 10, and Top 3.

```python
song_charts_temp = song_charts_unmod[:]
song_charts_temp['song'] = song_charts['song'].map(lambda x: x.replace('amp;', '').replace('\\', '').replace('&quot;', '\"').replace('&#039;','\''))
song_charts_temp['artist'] = song_charts['artist'].map(lambda x: x.replace('amp;', '').replace('\\', ',').replace('&quot;', '\"').replace('&#039;','\''))
song_charts_temp['featuring'] = song_charts['featuring'].map(lambda x: x.replace('amp;', '').replace('\\', ',').replace('&quot;', '\"').replace('&#039;','\'') if pd.notnull(x) else float('nan'))
song_charts_temp = song_charts_temp.drop(['duration', 'peak'], axis=1)
everything = pd.merge(combo, song_charts_temp, on=['song', 'artist'])
everything['score'] = everything.apply(lambda r: (r.duration + 10*(r.duration/50) + (100-r.peak) + 50*(4 - r.peak/50 - int(r.peak > 10) - int (r.peak > 3))), axis=1)
everything
print everything.columns[:80]
print everything.columns[80:]
```

```python
target = everything['score'].as_matrix()
everything = everything.drop(['song','artist','debut', 'peak', 'num_weeks', 'longitude', 'latitude', 'duration', 'peak_week', 'total_artists', 'featuring_x', 'has_feature', 'time_signature', 'key'] + [str(week_num) for week_num in range(1,88)] + ['featuring_y', 'location','featuring'], axis=1)
everything.fillna(value=0, inplace=True)
data = everything.drop('score', 1).as_matrix()
print data
data_normalized = StandardScaler().fit_transform(data)
#data_normalized = data
features = everything.columns[:-1]
print target.shape
print data.shape
X = data_normalized
y = target
```
