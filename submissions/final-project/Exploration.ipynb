## Data Collection

In order to gather chart, song, and lyrics data for my project, I scraped Billboard.com, used Echo Nest's music API, and scraped Genius.com. The code for the scraping is included in my project folder. I decided to scrape in Terminal, as I find it easier to debug and write long pieces of code there instead of iPython.

The first step was scraping Billboard, which I did in three steps: downloading the charts (get_charts.py), scraping the songs (song_scrape.py), and then calculating the peak week and position for each song afterwards (peak.py). I chose to focus on chart data starting from 2012, because looking at an extended period of time made for a very large dataset and even with 3.5 years of songs, I already had over 1300+ records to look at across each of my three sets of data. 

Next, I used the Echo Nest API to get its audio_summary for each song (attributes.py). Lastly, I searched and scraped for lyrics from Genius.com (get_lyrics.py), which has an accurate repository of lyrics for most popular songs.

All of these files are saved as CSVs, so I could easily retrieve them. When loaded, the unicode encoding had to be corrected to properly display some symbols in the song or artist names. Songs that were performed by different artists have (artist name) appended to the song title. There are a total of 87 columns as the maximum chart duration of all the songs is an impressive 87 weeks for Radioactive by Imagine Dragons.

```python
def evaluate_clusters(X,max_clusters):
    error = np.zeros(max_clusters+1)
    error[0] = 0;
    for k in range(1,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(X)
        error[k] = kmeans.inertia_

    plt.plot(range(1,len(error)),error[1:])
    plt.xlabel('Number of clusters')
    plt.ylabel('Error')
```

```python
def analyze_lyrics(song, artist):
    lyrics = []
    filename = os.path.join('data/lyrics/' + song.replace('/','') + '-' + artist + '.csv')
    if os.path.isfile(filename):
        f = open(filename)
        raw = f.read().lower().replace(',', '').replace('feat.','').split(';')

        for line in raw:
            lyrics += line.split()

        lyrics_text = nltk.Text(lyrics)
        lyrics_blob = TextBlob(' '.join(lyrics))

        unique = len(set(lyrics_text))
        uniqueness = unique/(len(lyrics_text)*1.0)
        #print "Unique: " + str(unique)
        #print "Uniqueness: " + str(uniqueness)
        fdist1 = nltk.FreqDist(lyrics_text)
        top_words = fdist1.most_common(10)
        #print top_words
        sentiment = lyrics_blob.sentiment.polarity
        happy = False
        if sentiment > 0:
            happy = True
        #print type(sentiment)
        #print sentiment
        return {'unique': unique, 'uniqueness': uniqueness, 'top_words': top_words, 'polarity': sentiment, 'happy': happy}
```

```python
stemmer = SnowballStemmer("english")
def tokenize_and_stem(text):
    # first tokenize by sentence, then by word to ensure that punctuation is caught as it's own token
    tokens = [word for sent in nltk.sent_tokenize(text) for word in nltk.word_tokenize(sent)]
    filtered_tokens = []
    # filter out any tokens not containing letters (e.g., numeric tokens, raw punctuation)
    for token in tokens:
        if re.search('[a-zA-Z]', token):
            filtered_tokens.append(token)
    stems = [stemmer.stem(t) for t in filtered_tokens]
    return stems


def tokenize_only(text):
    # first tokenize by sentence, then by word to ensure that punctuation is caught as it's own token
    tokens = [word.lower() for sent in nltk.sent_tokenize(text) for word in nltk.word_tokenize(sent)]
    filtered_tokens = []
    # filter out any tokens not containing letters (e.g., numeric tokens, raw punctuation)
    for token in tokens:
        if re.search('[a-zA-Z]', token):
            filtered_tokens.append(token)
    return filtered_tokens

totalvocab_stemmed = []
totalvocab_tokenized = []
for i in lyrics:
    allwords_stemmed = tokenize_and_stem(i) #for each item in 'synopses', tokenize/stem
    totalvocab_stemmed.extend(allwords_stemmed) #extend the 'totalvocab_stemmed' list
    
    allwords_tokenized = tokenize_only(i)
    totalvocab_tokenized.extend(allwords_tokenized)
    
vocab_frame = pd.DataFrame({'words': totalvocab_tokenized}, index = totalvocab_stemmed)
print 'there are ' + str(vocab_frame.shape[0]) + ' items in vocab_frame'
```
