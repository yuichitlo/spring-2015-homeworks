```python
import pandas as pd
from pandas.compat import u
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import sklearn.utils as utils
import sklearn.linear_model as linear_model
import sklearn.cross_validation as cross_validation
from sklearn.preprocessing import normalize
import sklearn.metrics as metrics
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
song_charts = pd.read_csv('data/chart-info-addt.csv', parse_dates=['Debut', 'peak_week'], encoding='latin-1')
print song_charts.head(1)
song_charts.columns = [label.lower() for label in song_charts.columns]
song_charts['song'] = song_charts['song'].map(lambda x: x.replace('amp;', '').replace('\\', '').replace('&quot;', '\"').replace('&#039;','\''))
song_charts['artist'] = song_charts['artist'].map(lambda x: x.replace('amp;', '').replace('\\', ',').replace('&quot;', '\"').replace('&#039;','\''))
song_charts['featuring'] = song_charts['featuring'].map(lambda x: x.replace('amp;', '').replace('\\', ',').replace('&quot;', '\"').replace('&#039;','\'') if pd.notnull(x) else float('nan'))

print len(song_charts)
print len(song_charts.columns)
print song_charts.info()
```

```python
song_charts.head()
```

```python
average_peak = song_charts['peak'].mean()
print 'Average peak: ' + str(average_peak)
average_duration = song_charts['duration'].mean()
print 'Average duration: ' + str(average_duration)
```

```python
min_peak = song_charts['peak'].min()
max_peak = song_charts['peak'].max()
print 'Peak\t\tMin: ' + str(min_peak) + '\t' + 'Max: ' + str(max_peak)
min_duration = song_charts['duration'].min()
max_duration = song_charts['duration'].max()
print 'Duration\tMin: ' + str(min_duration) + '\t' + 'Max: ' + str(max_duration)
```

```python
top_artists_count = song_charts[['song','artist']].groupby('artist').count().sort('song', ascending=False)
top_artists_avg_peak = song_charts[['song', 'artist', 'peak']].groupby('artist').mean().sort('peak', ascending=True)
top_artists = pd.concat([top_artists_count, top_artists_avg_peak], axis=1)
num_artists = len(top_artists)
print "Number of artists overall: " + str(num_artists)
num_top_artists = len(top_artists[top_artists.peak > 80])
print "Number of artists above average peak: " + str(num_top_artists)
```

```python
fig, ax = plt.subplots()
ax.scatter(top_artists.peak, top_artists.song, c="slategray", alpha=0.3, linewidths=0.2)
ax.set_xlabel('Average Peak Position')
ax.set_ylabel('Number of Songs')
ax.set_title('Average Peak Position vs Number of Charting Songs')
```

```python
song_charts['num_weeks'] = song_charts.count(axis=1)
song_charts['num_weeks'] = song_charts['num_weeks'].apply(lambda r: r - 6)
song_charts['has_feature'] = song_charts['featuring'].apply(lambda r: pd.notnull(r))
song_charts['num_features'] = song_charts['featuring'].apply(lambda r: len(r.replace('Featuring ','').replace(' & ',',').replace(' , ', ',').split(','))\
                                                             if pd.notnull(r) else 0)
song_charts
max_features = song_charts[song_charts['num_features'] == max(song_charts.num_features)]
max_features
```
