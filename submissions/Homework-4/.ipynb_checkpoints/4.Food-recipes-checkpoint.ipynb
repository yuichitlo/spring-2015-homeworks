## Data Parsing + Representation

My hypothesis is that cuisines with similar flavor profiles, ethnographic makeup, or geographic location will utilize similar ingredients in their recipes.

To explore this, I am using the recipe list from Epicurious. The other lists refer to country and more generic cuisine categorizing, so they are not quite compatible with this list. The basic data frame structure is that of Cuisine, Ingredient pairings.

```python
# Parsing the original tab-delimited text file into a dictionary which is then converted to a data frame
epic_recipes = {"Cuisine": [], "Ingredient": []}

with open('data/epic_recipes.txt') as f:
    for line in f:
        tokens = line.split('\t')
        for token in tokens[1:]:
            if '\n' in token:
                token = token.replace('\n', '')
            epic_recipes['Cuisine'] += [tokens[0]]
            epic_recipes['Ingredient'] += [token]
    f.close()

epic_recipes = pd.DataFrame(epic_recipes)
```

Once I have my data brought into a dataframe, I tried to look at the data and aggregate it in different ways. Grouping along the two columns and getting to know the data better.

I found that grouping by cuisines and then aggregating the number of recipes within each cuisine group that uses a specific ingredient will best reflect the question behind my hypothesis. Knowing the cuisines and ingredients, I established a mapping for the categories and then translated the foundational dataframe into a matrix that reflects the grouping I selected.

```python
# Number of recipes using each ingredient per cuisine
cuisine_ingr_count = epic_recipes.groupby(['Cuisine', 'Ingredient']).size()

# Number of unique ingredients per cuisine
cuisine_ingr_total = cuisine_ingr_count.reset_index().groupby(['Cuisine']).size()

# Number of recipes in each cuisine where an ingredient is used
#ingr_cuisine_count = epic_recipes.groupby(['Ingredient', 'Cuisine']).size()
#print ingr_cuisine_count.head()

# Total count of unique cuisines and unique ingredients
num_cuisines = len(epic_recipes.Cuisine.unique())
num_ingredients = len(epic_recipes.Ingredient.unique())

#
# Cuisine - Ingredient comparison matrix prep
#
cuisine_ingr_count = cuisine_ingr_count.reset_index()
cuisine_ingr_count.columns = ['Cuisine', 'Ingredient', 'Count']

# Mapping from 0-25 of cuisines in alphabetical order
cuisine_mapping = {}
cuisines = epic_recipes.Cuisine.unique()
cuisines.sort()
for cuisine in cuisines:
    cuisine_mapping[cuisine] = len(cuisine_mapping)
#print cuisine_mapping

# Mapping from 0-349 of ingredients in alphabetical order
ingr_mapping = {}
ingrs = epic_recipes.Ingredient.unique()
ingrs.sort()
for ingr in ingrs:
    ingr_mapping[ingr] = len(ingr_mapping)
    
# 26 x 350 array Frequency of ingredient used in recipes for each cuisine    
coded = np.zeros(shape=(len(cuisine_mapping),len(ingr_mapping)))

# 26 x 350 array Use (0,1) of ingredient in recipes for each cuisine
coded_bin = np.zeros(shape=(len(cuisine_mapping),len(ingr_mapping)))

# Fill in coded and coded_bin matrices
for row, data in cuisine_ingr_count.iterrows():
    coded[cuisine_mapping[data['Cuisine']]][ingr_mapping[data['Ingredient']]] = data['Count']
    yes = float(bool(data['Count']))
    coded_bin[cuisine_mapping[data['Cuisine']]][ingr_mapping[data['Ingredient']]] = yes

#print np.asmatrix(coded_bin[22])
```

## Clustering

Now that the data is prepped into a matrix of coded numerical values, I tried two ways of hierarchically clustering it. The outputs for Euclidean distance and Cosine similarity produced some interesting results.

The Euclidean distance clustering produced 12 different clusters, while the cosine similarity produced 10 different clusters. The Euclidean distance clustering had more singleton clusters that are essentially outliers, while most of the clusters produced by calculating cosine similarity as distance actually grouped cuisines together.

```python
# Hierarchically clustering data
#coded = StandardScaler().fit_transform(coded)
print "EUCLIDEAN DISTANCE"
Z = hr.linkage(coded, method='complete', metric='euclidean')
print Z.shape, coded.shape

distanceMatrix = pdist(coded,'euclidean')
ck = hr.fcluster(hr.linkage(distanceMatrix, method='complete'), 280,'distance')
print ck
ck_freq = {}
for label in ck:
    if label not in ck_freq:
        ck_freq[label] = 1
    else:
        ck_freq[label] += 1
print ck_freq
print

print "COSINE SIMILARITY"
Z_2 = hr.linkage(coded, method='complete', metric='cosine')
print Z_2.shape, coded.shape

distanceMatrix = pdist(coded,'cosine')
ck = hr.fcluster(hr.linkage(distanceMatrix, method='complete'), 0.18,'distance')
print ck
ck_freq = {}
for label in ck:
    if label not in ck_freq:
        ck_freq[label] = 1
    else:
        ck_freq[label] += 1
print ck_freq
```

## Visualizing

Using a dendogram, I visualized my clusterings from the two different distance functions used above. The different clusterings required different color_thresholds to better pinpoint the groupings between cuisines with similar ingredients.

The issue with singleton clusters comes into play in the first dendrogram; many of the cuisines were not merging together until the threshold grew to a very large distance. In the second graph however, we can see that most clusters grouped early on, indicating a clearer gradual hierarchy between very similar, similar, and distantly similar cuisines.

We can see that regional relationships between cuisines inform many of the clusters; Thai, Vietnamese, Japanese, Asian, Chinese; Cajun/Creole, Southwestern, Central/South American, Mexican; Indian, Middle Eastern, African, Moroccan; Greek, Mediterranean, Italian, Spanish/Portuguese; German, Scandanavian; Jewish, Eastern European/Russian, American, French; Southern/Soul Food, English/Scottish, Irish.

Generically, the clusters could be summed up as, Asian (Southeast, Northeast), Latin flavors, spicy, Central European, German, and the two last clusters that are a bit more difficult to sum up succinctly. I will be exploring the last cluster in the next section.

## Analysis of Specific Cluster

The combination of English/Scottish and Irish cuisine reflects geographic proximity, however the connection of Southern/Soul Food to these two cuisines seems like a mistake at first glance.

Once we review their most frequently used ingredients however, we see why these three are placed in a cluster. The top 3 ingredients in exact order for all three of these cuisines are butter, wheat, and eggs. These cuisines are all rich and very liberal with their usage of fat. It is especially intriguing to see that nearly the entire top 5 of English and Irish cuisine are the same: butter, wheat, eggs, cream, and milk/buttermilk.

In the dendrogram, Southern/Soul Food did not merge with English and Irish food for awhile and we can see that it is largely due to the more diverse range of ingredients used in this cuisine. The scatter plot of all the ingredients and frequencies display many points shared by English and Irish cuisine, while soul food ingredients often ranking in higher frequency and lack of use by the other two cuisines. The column inhabited by a few of the soul food points consists of no other dots, meaning it is only used in soul food amongst the three cuisines in this cluster. All the ingredient counts for soul food skews higher than the two other cuisines indicating more soul food recipes were probably in the original dataset. This contributes to its increased variety, though key differences in ingredients that we assume between the hearty, flavorful food of the South and rustic, earthy food of the UK make up the main reason why these are different.

```python
soul = cuisine_ingr_count[cuisine_ingr_count.Cuisine == 'Southern_SoulFood']
soul.sort(columns=['Count'], ascending=False, inplace=True)
print soul.head()
print

english = cuisine_ingr_count[cuisine_ingr_count.Cuisine == 'English_Scottish']
english.sort(columns=['Count'], ascending=False, inplace=True)
print english.head()
print

irish = cuisine_ingr_count[cuisine_ingr_count.Cuisine == 'Irish']
irish.sort(columns=['Count'], ascending=False, inplace=True)
print irish.head()
print

cluster = soul.merge(english, on='Ingredient', how='outer').merge(irish, on='Ingredient', how='outer')
cluster.fillna(value=0, inplace=True)
cluster.drop(['Cuisine_x', 'Cuisine_y', 'Cuisine'], 1, inplace=True)
cluster.columns = ['Ingredient', 'Soul', 'English', 'Irish']
print cluster

# Create matrices of ingredients and counts for each cuisine
cluster_soul = np.zeros(shape=(len(cluster.Ingredient.unique()), 2))
cluster_english = np.zeros(shape=(len(cluster.Ingredient.unique()), 2))
cluster_irish = np.zeros(shape=(len(cluster.Ingredient.unique()), 2))
cluster_ingr_id = []
cluster_ingr_name = []
ingr_all = 0

for row, data in cluster.iterrows():
    ingr_code = ingr_mapping[data['Ingredient']]
    cluster_ingr_id += [ingr_code]
    cluster_ingr_name += data['Ingredient']
    if data['Soul'] > 10:
        cluster_soul[row][0] = ingr_code
        cluster_soul[row][1] = data['Soul']
        
    if data['English'] > 10:
        cluster_english[row][0] = ingr_code
        cluster_english[row][1] = data['English']
    
    if data['Irish'] > 10:
        cluster_irish[row][0] = ingr_code
        cluster_irish[row][1] = data['Irish']
    
    if data['Soul'] > 0 and data['English'] > 0 and data['Irish'] > 0:
        ingr_all += 1
        
# Statistics analyzing the disparity between ingredient lists of the three cuisines
print ingr_all, ingr_all/(len(cluster)*1.0)
print len(soul)
print len(english)
print len(irish)
    
indian = cuisine_ingr_count[cuisine_ingr_count.Cuisine == 'Indian']
indian.sort(columns=['Count'], ascending=False, inplace=True)
print indian.head()
print

african = cuisine_ingr_count[cuisine_ingr_count.Cuisine == 'African']
african.sort(columns=['Count'], ascending=False, inplace=True)
print african.head()
```

```python
fig, ax = plt.subplots()
plot_soul = mpatches.Patch(color='b', label='Soul')
ax.plot(cluster_soul[:,0], cluster_soul[:,1],'bo')
plot_english = mpatches.Patch(color='r', label='English')
ax.plot(cluster_english[:,0], cluster_english[:,1],'ro')
plot_irish = mpatches.Patch(color='g', label='Irish')
ax.plot(cluster_irish[:,0], cluster_irish[:,1],'go')
plt.xlabel('Ingredient')
plt.ylabel('Frequency')
plt.title('Ingredient Use per Cuisine')
plt.legend(handles = [plot_soul, plot_english, plot_irish], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.tight_layout()
```
