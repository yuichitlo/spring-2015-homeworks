```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import scipy.cluster.hierarchy as hr
from sklearn.preprocessing import StandardScaler
from scipy.spatial.distance import pdist, squareform
import sklearn.metrics as metrics

#import matplotlib as mpl
import seaborn as sns
%matplotlib inline
```

```python
epic_recipes = {"Cuisine": [], "Ingredient": []}

with open('data/epic_recipes.txt') as f:
    for line in f:
        tokens = line.split('\t')
        for token in tokens[1:]:
            if '\n' in token:
                token = token.replace('\n', '')
            epic_recipes['Cuisine'] += [tokens[0]]
            epic_recipes['Ingredient'] += [token]
    f.close()

epic_recipes = pd.DataFrame(epic_recipes)
```

```python
# Number of recipes using each ingredient per cuisine
cuisine_ingr_count = epic_recipes.groupby(['Cuisine', 'Ingredient']).size()

# Number of unique ingredients per cuisine
cuisine_ingr_total = cuisine_ingr_count.reset_index().groupby(['Cuisine']).size()

# Number of recipes in each cuisine where an ingredient is used
#ingr_cuisine_count = epic_recipes.groupby(['Ingredient', 'Cuisine']).size()
#print ingr_cuisine_count.head()

# Total count of unique cuisines and unique ingredients
num_cuisines = len(epic_recipes.Cuisine.unique())
num_ingredients = len(epic_recipes.Ingredient.unique())

#
# Cuisine - Ingredient comparison matrix prep
#
cuisine_ingr_count = cuisine_ingr_count.reset_index()
cuisine_ingr_count.columns = ['Cuisine', 'Ingredient', 'Count']

# Mapping from 0-25 of cuisines in alphabetical order
cuisine_mapping = {}
cuisines = epic_recipes.Cuisine.unique()
cuisines.sort()
for cuisine in cuisines:
    cuisine_mapping[cuisine] = len(cuisine_mapping)
#print cuisine_mapping

# Mapping from 0-349 of ingredients in alphabetical order
ingr_mapping = {}
ingrs = epic_recipes.Ingredient.unique()
ingrs.sort()
for ingr in ingrs:
    ingr_mapping[ingr] = len(ingr_mapping)
    
# 26 x 350 array Frequency of ingredient used in recipes for each cuisine    
coded = np.zeros(shape=(len(cuisine_mapping),len(ingr_mapping)))

# 26 x 350 array Use (0,1) of ingredient in recipes for each cuisine
coded_bin = np.zeros(shape=(len(cuisine_mapping),len(ingr_mapping)))

# Fill in coded and coded_bin matrices
for row, data in cuisine_ingr_count.iterrows():
    coded[cuisine_mapping[data['Cuisine']]][ingr_mapping[data['Ingredient']]] = data['Count']
    yes = float(bool(data['Count']))
    coded_bin[cuisine_mapping[data['Cuisine']]][ingr_mapping[data['Ingredient']]] = yes

#print np.asmatrix(coded_bin[22])
```

```python
mdata = epic_recipes[['Cuisine', 'Ingredient']]
mdata_labels = pd.Categorical.from_array(mdata['Cuisine'])
mdata['cuisine_code'] = mdata_labels.codes
mdata_labels = pd.Categorical.from_array(mdata['Ingredient'])
mdata['ingr_code'] = mdata_labels.codes
mdata.fillna(value=0, inplace=True)
mdata['cuisine_code'] = mdata['cuisine_code'].apply(lambda x: x*1.0)
mdata['ingr_code'] = mdata['ingr_code'].apply(lambda x: x*1.0)
print mdata.head()
X = StandardScaler().fit_transform(mdata[['cuisine_code', 'ingr_code']])
np.random.shuffle(X)
X_sub = X[:100]
#print X_sub
print X.shape
print X_sub.shape
```

```python
mdata_scaled = pd.DataFrame(X)
mdata_scaled.columns = ['s_cuisine_code', 's_ingr_code']
print mdata_scaled.head()
print mdata.head()
mdata_merge = pd.concat([mdata, mdata_scaled], axis=1)
print mdata_merge.tail()
```

```python
mdata_sub = np.array(mdata_merge)
np.random.shuffle(mdata_sub)
#print mdata_sub[:10]
mdata_sub_coords = mdata_sub[:][4:]
print mdata_sub_coords
```

```python
#coded = StandardScaler().fit_transform(coded)
Z = hr.linkage(coded, method='complete', metric='euclidean')
print Z.shape, coded.shape
fig = plt.figure(figsize=(10,10))
#Dendogram is reordered by the dendogram command to improve the way it looks
T = hr.dendrogram(Z,color_threshold=280, leaf_font_size=12, labels=cuisines, leaf_rotation=90) #Same color for anything above 0.4 on y-axis
fig.show()
```

```python
distanceMatrix = pdist(coded,'euclidean')
ck = hr.fcluster(hr.linkage(distanceMatrix, method='complete'), 280,'distance')
print ck
```

```python
distances = metrics.euclidean_distances(coded)
cg = sns.clustermap(distances, method="complete", figsize=(13,13), xticklabels=False) #method=complete runs a hierarchical clustering alg
print cg.dendrogram_col.reordered_ind
```

```python
coded_bin = StandardScaler().fit_transform(coded_bin)
Z_bin = hr.linkage(coded_bin, method='complete', metric='euclidean')
print Z.shape, coded.shape
fig_bin = plt.figure(figsize=(10,5))
#Dendogram is reordered by the dendogram command to improve the way it looks
T_bin = hr.dendrogram(Z_bin, color_threshold=25, leaf_font_size=12, labels=cuisines, leaf_rotation=90) #Same color for anything above 0.4 on y-axis
fig_bin.show()
```
